```python
import json
import pprint

with open('pic1.json') as f:
  data = json.load(f)

pprint.pprint(data)
```

```python
def findCorners(points):
    left, top = 2147483646, 2147483646
    right, bottom = 0, 0
    for feat, coord in points.items():
        if coord['x'] < left: 
            left = coord['x']
        if coord['x'] > right: 
            right = coord['x']
        if coord['y'] < top: 
            top = coord['y']
        if coord['y'] > bottom: 
            bottom = coord['y']
            
    return left, right, bottom, top
        
```

```python
landmarks = data['faces'][0]['landmark']

nose_keys = [
    'nose_contour_left1', 
    'nose_contour_left2', 
    'nose_contour_left3', 
    'nose_contour_lower_middle', 
    'nose_contour_right1', 
    'nose_contour_right2',
    'nose_contour_right3',
    'nose_left',
    'nose_right',
    'nose_tip'
]

nose_dict = {k:v for k,v in landmarks.items() if k in nose_keys}

left, right, bottom, top = findCorners(nose_dict)
print(left, right, bottom, top)
```

```python
img = cv2.imread('pic1.jpg')
img[top:bottom, left:right]=0
mask = img.copy()
mask[:,:] = 0
mask[top:bottom, left:right]=255
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
dst = cv2.inpaint(img,mask,25,cv2.INPAINT_TELEA)
cv2.imshow('dst',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
